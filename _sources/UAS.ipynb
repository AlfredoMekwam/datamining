```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HamedAyani114/dataset/main/winequality-red.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.quality.value_counts()
```

```python
df['quality'] = (df['quality'] >= 6).astype(int) #jika quality lebihbesar samadengan 6 maka quality=1
df.head()
```

```python
# scaler = StandardScaler()
scaler = MinMaxScaler()
```

```python
dfC=df.drop(columns=['quality']) #data tanpa quality
dfL= pd.DataFrame(df, columns = ['quality']) #data quality
```

```python
dfC
```

```python
dfL
```

```python
DataNorm=scaler.fit_transform(dfC.values)
```

```python
frameDataNorm=pd.DataFrame(DataNorm)
frameDataNorm.columns=['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol']
```

```python
df_new = pd.concat([frameDataNorm, dfL], axis=1)
df_new
```

```python

X = df_new.iloc[:,:-1].values 
y = df_new.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.30, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_test)
print(classification_report(y_test,rf_pred))
```

```python
model_accuracy = pd.DataFrame(columns=['Model','Accuracy'])
models = {
          "KNN" : KNeighborsClassifier(n_neighbors=7),
          "DT" : DecisionTreeClassifier(),
          'RFC' : RandomForestClassifier(n_estimators=100),
          "GNB" : GaussianNB(),

          }

for test, clf in models.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    # acc = accuracy_score(y_test,y_pred)
    train_pred = clf.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    print("\n", test + ' scores')
    # print(acc)
    print(classification_report(y_test,y_pred))
    print('*' * 100,"\n")
    model_accuracy = model_accuracy.append({'Model': test, 'Train_acc': train_acc}, ignore_index=True)
```

```python
# model
# from sklearn import model_selection
import pickle
# save the model to disk
model=RandomForestClassifier(n_estimators=100)
filename = 'modelwineRFC.sav'
model.fit(X_train,y_train)
model_pred = model.predict(X_test)
print(classification_report(y_test,model_pred))

pickle.dump(model, open(filename, 'wb'))
```
