1. Importing Dataset:

```python
dataset = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
```

2. Getting Info About Dataset

```python
dataset.head()
dataset.info()
dataset = dataset.drop(["id"], axis = 1)
dataset = dataset.drop(["Unnamed: 32"], axis = 1)
dataset.head(3)
M = dataset[dataset.diagnosis == "M"]
M.head(5)
B = dataset[dataset.diagnosis == "B"]
B.head(5)
```

3. Dataset Visualization

```python
plt.title("Malignant vs Benign Tumor")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.scatter(M.radius_mean, M.texture_mean, color = "red", label = "Malignant", alpha = 0.3)
plt.scatter(B.radius_mean, B.texture_mean, color = "lime", label = "Benign", alpha = 0.3)
plt.legend()
plt.show()
```

4. Meaning Of Decision Tree Algorithm
Model pohon keputusan di mana variabel target menggunakan seperangkat nilai diskrit diklasifikasikan sebagai Pohon Klasifikasi.
Di pohon-pohon ini, setiap node, atau daun, mewakili label kelas sementara cabang mewakili konjungsi fitur yang mengarah ke label kelas.
Pohon keputusan di mana variabel target mengambil nilai kontinu, biasanya angka, disebut Pohon Regresi.
Kedua tipe tersebut biasa disebut bersama pada CART (Classification and Regression Tree).


5. Decision Tree with Sklearn

```python
dataset.diagnosis = [1 if i == "M" else 0 for i in dataset.diagnosis]
x = dataset.drop(["diagnosis"], axis = 1)
y = dataset.diagnosis.values
# Normalization:
x = (x - np.min(x)) / (np.max(x) - np.min(x))
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
DecisionTreeClassifier()
# prediction
dt.score(x_test, y_test)
```
